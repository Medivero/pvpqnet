```python

import numpy as np
import matplotlib.pyplot as plt
import math as Math


# generate mumpy array with 5000 elements from 1 to 5000
x = np.linspace(1, 5000, 5000)
# generate mumpy array y using a custom function taking an element of x as input

def ladderScore(x):
    if x>=1 and x<=50:
        return 600 + (50 - x) * (400 / 50)
    elif x>=51 and x<=100:
        return 400 + (100 - x) * (200 / 50)
    elif x>=101 and x<=5000:
        return 0 + round((5000 - x) * (400 / 4900))
    else:
        return 0
    
ladderScroreNp = np.frompyfunc(ladderScore, 1, 1)
y = ladderScroreNp(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('')
plt.grid(True)
plt.show()
```
